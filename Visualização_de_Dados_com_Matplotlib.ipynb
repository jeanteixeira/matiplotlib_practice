**Importando biblioteca**

```
import matplotlib.pyplot as plt
```

**Definindo variáveis**

```
x      = [1, 3, 5, 7, 9]
y      = [2, 3, 7, 1, 0]

x2     = [2, 4, 6, 8, 10]
y2     = [5, 1, 3, 7, 4]

eixo_x = "Eixo X"
eixo_y = "Eixo_Y"
```

## **Gráfico de Linha**

```
plt.title("Gráfico de Linha")
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

plt.plot(x, y)
plt.show()
```

## **Gráfico de Barras**

```
plt.title("Gráfico de Barras")
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

plt.bar(x, y)
plt.show()
```

## **Gráfico de Barras - Comparativo**

```
plt.title("Gráfico de Barras - Comparativo")
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

plt.bar(x, y, label = "Grupo 1")
plt.bar(x2, y2, label = "Grupo 2")
plt.legend()
plt.show()
```

## **Gráfico de Dispersão**

```
plt.title("Gráfico de Dispersão")
plt.xlabel(eixo_x)
plt.ylabel(eixo_y)

plt.scatter(x, y, label = "meus pontos", color = "r", marker = "x", s = 60)
plt.plot(x, y, color = "k", linestyle = ":")
plt.legend()
plt.show()
```

```
dados = open("/content/populacao_brasileira.csv").readlines()

x = []
y = []

for i in range(len(dados)):
  if i!= 0:
    linha = dados[i].split(";")
    x.append(int(linha[0]))
    y.append(int(linha[1]))

plt.title("Crescimento da População Brasileira de 1980-2016")
plt.xlabel("Ano")
plt.ylabel("População x100.000.000")

plt.plot(x, y, color = "k", linestyle = "--")
plt.bar(x, y, color = "#ccc")
plt.show()
```
